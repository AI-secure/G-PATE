```python
import pickle
import numpy as np
import matplotlib.pyplot as plt
from keras.layers.pooling import MaxPooling2D
from keras.layers.convolutional import Convolution2D, Conv2D

from keras.datasets import mnist
from keras.models import Sequential
from keras.layers.core import Dense, Dropout, Activation, Flatten
from keras.utils import np_utils
from keras.callbacks import TensorBoard, EarlyStopping
from keras import optimizers
# Choose the class
nb_classes = 10
im_rows = 28
im_cols = 28
batch_size = 512
im_shape = (im_rows, im_cols, 1)
class1 = 2
class2 = 3
```

```python
# dir to sythetic data
datafile = 'samples/private.data'
with open(datafile,'rb') as f:
    data = pickle.load(f)
```

```python
# Create 20 subplots
fig, axes = plt.subplots(2, 10, figsize=(10, 2))

for i in range(2):
    for j in range(10):
        data_sel = data[data[:,-10+j]==1]
        idx = np.random.randint(0, 6000)
#         print("%d, %d, %d" % (i,j,idx))
        axes[i][j].imshow(data_sel[idx,:784].reshape(28, 28), cmap='gray')
        axes[i][j].set_xticks([])
        axes[i][j].set_yticks([])
```

```python
if nb_classes == 2:
    data_class1 = data[data[:,-10+class1]==1]
    data_class2 = data[data[:,-10+class2]==1]
    data = np.concatenate((data_class1,data_class2),axis=0)
    x, label = np.hsplit(data,[-10])
    label_class1 = label[:,class1]
    label_class2 = label[:,class2]
    label = np.concatenate((label_class1,label_class2),axis=0)
else:
    x, label = np.hsplit(data,[-10])
    
label = label.reshape((label.shape[0],nb_classes),order='F')
x = x.reshape(x.shape[0], 28, 28, 1)
    
# deal with mnist
(x_train, y_train), (x_test, y_test) = mnist.load_data()
y_test = np_utils.to_categorical(y_test, 10)

if nb_classes == 2:
    idx = y_test[:,class1 or class2] ==1
    y_test = y_test[idx,:]
    x_test = x_test[idx,:,:]
    y_test_class1 = y_test[:,class1]
    y_test_class2 = y_test[:,class2]
    y_test = np.concatenate((y_test_class1,y_test_class2),axis=0)
    
y_test = y_test.reshape((x_test.shape[0],nb_classes),order='F')
x_test = x_test.reshape(x_test.shape[0], 28, 28, 1)
x_test = x_test.astype('float32') / 255.
```

```python
shuffle_idx = np.random.permutation(np.arange(len(x)))
x = x[shuffle_idx]
label = label[shuffle_idx]
```

```python
model = Sequential()
model.add(Conv2D(32, kernel_size=3, activation='relu', input_shape=im_shape, name='Conv2D-1'))
model.add(MaxPooling2D(pool_size=2, name='MaxPool'))
model.add(Dropout(0.2, name='Dropout-1'))
model.add(Conv2D(64, kernel_size=3, activation='relu', name='Conv2D-2'))
model.add(Dropout(0.25, name='Dropout-2'))
model.add(Flatten(name='flatten'))
model.add(Dense(64, activation='relu', name='Dense'))
model.add(Dense(nb_classes, activation='softmax', name='Output'))
sgd = optimizers.sgd(lr=0.1, decay=1e-6, momentum=0.9, nesterov=True)

model.compile(loss='categorical_crossentropy', optimizer='sgd', metrics=['accuracy'])

tbCallBack = TensorBoard(log_dir='./logs',  
                 histogram_freq=0,
                 write_graph=True,
                 write_grads=True,
                 write_images=True,
                 embeddings_freq=0,
                 embeddings_layer_names=None,
                 embeddings_metadata=None)


model.fit(x, label,
          batch_size=batch_size, nb_epoch=20,
          verbose=1,
          validation_data=(x_test, y_test),
          callbacks=[tbCallBack])
score = model.evaluate(x_test, y_test, verbose=0)
```
